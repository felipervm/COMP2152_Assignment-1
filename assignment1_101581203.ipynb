````python
# Step 3: Create 4 variables with their data types
gym_member = "Alex Alliton"  # str
preferred_weight_kg = 20.5  # float
highest_reps = 25  # int
membership_active = True  # bool
````

````python
# Step 4: Create a dictionary with friends' workout data
# Dictionary with string keys and tuple values (dict)
workout_stats = {
    "Alex": (30, 45, 20),  # (yoga, running, weightlifting)
    "Jamie": (40, 30, 50),
    "Taylor": (25, 60, 35)
}

print("Original workout_stats:")
print(workout_stats)
````

````python
# Step 5: Calculate total workout minutes for each friend using a loop
for friend, minutes in workout_stats.items():
    # Only process if it's not already a total key
    if not friend.endswith("_Total"):
        total_minutes = sum(minutes)
        workout_stats[f"{friend}_Total"] = total_minutes

print("\nWorkout stats with totals:")
print(workout_stats)
````

````python
# Step 6: Create a 2D list from the dictionary
# 2-dimensional list (nested list) - list
workout_list = []
for friend, minutes in workout_stats.items():
    # Only add entries that are tuples (not the total values)
    if isinstance(minutes, tuple):
        workout_list.append(list(minutes))

print("\nWorkout list (2D):")
print(workout_list)
````

````python
# Step 7: Slice the workout_list
# Extract yoga and running minutes (first two columns) for all friends
print("\nYoga and Running minutes for all friends:")
for i, row in enumerate(workout_list):
    yoga_running = row[0:2]  # Slice first two elements
    print(f"Friend {i+1}: {yoga_running}")

# Extract weightlifting minutes (third column) for the last two friends
print("\nWeightlifting minutes for the last two friends:")
last_two_weightlifting = [row[2] for row in workout_list[-2:]]
print(last_two_weightlifting)
````

````python
# Step 8: Use if-statement within a loop to check total minutes
print("\nChecking for friends with >= 120 total minutes:")
for friend, value in workout_stats.items():
    # Check if this is a total key
    if friend.endswith("_Total"):
        if value >= 120:
            friend_name = friend.replace("_Total", "")
            print(f"Great job staying active, {friend_name}!")
````

````python
# Step 9: Allow user to input a friend's name and check if it exists
user_input = input("Enter a friend's name: ")

if user_input in workout_stats and isinstance(workout_stats[user_input], tuple):
    print(f"\n{user_input}'s workout minutes:")
    print(f"  Yoga: {workout_stats[user_input][0]} minutes")
    print(f"  Running: {workout_stats[user_input][1]} minutes")
    print(f"  Weightlifting: {workout_stats[user_input][2]} minutes")
    print(f"  Total: {workout_stats[user_input + '_Total']} minutes")
else:
    print(f"Friend {user_input} not found in the records.")
````

````python
# Step 10: Print friend with highest and lowest total workout minutes
# Get all totals
totals = {}
for friend, value in workout_stats.items():
    if friend.endswith("_Total"):
        friend_name = friend.replace("_Total", "")
        totals[friend_name] = value

# Find highest and lowest
highest_friend = max(totals, key=totals.get)
lowest_friend = min(totals, key=totals.get)

print("\n" + "="*50)
print("WORKOUT SUMMARY")
print("="*50)
print(f"Friend with highest total workout minutes: {highest_friend} ({totals[highest_friend]} minutes)")
print(f"Friend with lowest total workout minutes: {lowest_friend} ({totals[lowest_friend]} minutes)")
print("="*50)
````
